```python
#Now droping the Span column

df.drop(['Span'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# In the player name we are seeing the country name so removing it

df['Player'].str.split(pat='(').str[0]
```

```python
#Assigning this to the new column name (Player_Name)

df['Player_Name']=df['Player'].str.split(pat='(').str[0]
```

```python
# Now need to create the country column (We can use the value which we splitted from the Player column)

df['Player'].str.split(pat='(').str[1]
```

```python
df['Country']=df['Player'].str.split(pat='(').str[1]
```

```python
df['Country']=df['Country'].str.split(pat=')').str[0]
```

```python
#Now droping the "Player" Columns (Name+country)

df.drop(['Player'],axis=1,inplace=True)
```

```python
df = df[[df.columns[-2]] + list(df.columns[:-2]) + [df.columns[-1]]] #Changed the column to 1st row
```

```python
df.dtypes
```

```python
# Removing '*' from the Highest_Inns_Score

df['Highest_Inns_Score']=df['Highest_Inns_Score'].str.split(pat='*').str[0]
```

```python
df
```

```python

df['Highest_Inns_Score'].dtypes # Currently its in Objest
```

```python
# Now changing the data type

df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype(int)
```

```python
df['Highest_Inns_Score'].dtypes
```

```python
df.rename(columns={'Debut_Innings': 'Debut_Year'}, inplace=True)
df.head()
```

```python
# Now changing the data types for the (Debut_Innings, Last_year)

df=df.astype({'Debut_Year':'int','Last_year':'int'})
```

```python
df.dtypes
```

```python
# Formating Matches (Removing *)

df['Matches']=df['Matches'].str.split(pat='*').str[0]
```

```python
# Now converting the 'Matches' dataType from Object to Int

df['Matches']=df['Matches'].astype(int)
```

```python
df.dtypes
```

```python
#Now formating the 4s and 6s

df['4s']=df['4s'].str.split('+').str[0]
```

```python
df['6s']=df['6s'].str.split('+').str[0]
```

```python
# Changing the dtypes of 4s and 6s from Object to Int

df=df.astype({'4s':'int','6s':'int'})
```

```python
df.dtypes
```

```python
# Formating Ballfaced

df['Ball_Faced']=df['Ball_Faced'].str.split(pat='+').str[0]
```

```python
df['Ball_Faced']=df['Ball_Faced'].str.split(pat='-').str[0]
```

```python
# Question-2: Average career length

df['Career_Length'].mean()
```
